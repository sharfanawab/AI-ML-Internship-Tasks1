Task 1: Exploring and Visualizing Iris Dataset

## Objective
To explore the Iris dataset and visualize feature relationships.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
df.head()

df.shape

df.info()

df.describe()

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")
plt.title("Scatter Plot")
plt.show()
```

```python
df.hist(figsize=(8,6))
plt.suptitle("Histogram")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df)
plt.title("Box Plot")
plt.xticks(rotation=45)
plt.show()
```

```python
# Findings

1. Dataset contains 150 rows and 5 columns.
2. Three species are present:
   - Setosa
   - Versicolor
   - Virginica
3. Scatter plot shows clear separation among species.
4. Histograms show feature distributions.
5. Box plots help identify outliers.
```
